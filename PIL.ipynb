```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img=Image.open('image.jpeg')
img
```

# image resize

```python
w,h=img.size

img_resize=img.resize((w//2,h//2))
plt.imshow(img_resize)
plt.show()

img_resize=img.resize((w*2, h*2))
plt.imshow(img_resize)
plt.show()
```

```python
def two_image_show(im1, im2):
    plt.subplot(1,2,1)
    plt.imshow(im1)
    plt.subplot(1,2,2)    
    plt.imshow(im2)
    plt.show()
    
img_crop=img.crop((w//4,h//4,w*3//4,h*3//4))
# (left,upper,right,lower)

plt.imshow(img_crop)

two_image_show(img,img_crop)
```

# image rotate

```python
img_rotate=img.rotate(45)
# 도에 따라 img를 rotate

plt.imshow(img_rotate)
plt.show()
```

```python
img_flip=img.transpose(Image.FLIP_LEFT_RIGHT)
# 이미지 데이터를 상하좌우 FLIP 할 수 있음.

two_image_show(img,img_flip)
```

```python
img_flip_rotate=img.rotate(45).transpose(Image.FLIP_LEFT_RIGHT)
img_flip_rotate.save('save_img.png')
```
